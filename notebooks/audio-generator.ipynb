```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

from acoustic_no.utils.audio_utils import generate_audio_clip
```

```python
np.random.seed(231)
```

```python
# Random frequencies.
SIZE = 12
# Audio length
T = 0.05
num_audio = 20
# Sample rate
N = 44100
dt = 1 / N

# Sample random audio.
x = np.array([])
y = np.array([])
for i in range(num_audio):
    freqs = np.random.randint(20, 2000, size=(SIZE,))
    amps = np.random.rand(SIZE)
    x_clip, y_clip = generate_audio_clip(zip(freqs, amps), N, T)
    x = np.concatenate((x, T * i + x_clip), axis=0)
    y = np.concatenate((y, y_clip), axis=0)
# Normalize the sample
y = y / np.max(np.abs(y))
```

```python
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
wavfile.write("audio.wav", N, y)
```
